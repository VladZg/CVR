```python
import cv2
import matplotlib.pyplot as plt

cv2.namedWindow("thresholds")
cv2.createTrackbar("high_blue", "thresholds", 155, 255, lambda i : i)

# video_path = "./dotswhitenoise.mp4"
video_path = 0
cam = cv2.VideoCapture(video_path)

i = 0

while (True):
    success, frame = cam.read()

    if (success == False):
        cam.release()
        cam = cv2.VideoCapture(video_path)
        continue
    
    # frame[100:400, 300:500, :] = 30
    # i += 1

    # plt.hist(list(frame[::40, ::40, 0].flatten()), bins=40)
    # plt.show()

    hb = cv2.getTrackbarPos("high_blue", "thresholds")
    # print(hb)
    mask = cv2.inRange(frame, (0,0,0), (hb,255,255))

    # связность (количество и тип пикселей рядом)
    connectivity = 4
    output = cv2.connectedComponentsWithStats(mask, connectivity, cv2.CV_32S)
    #результат этой функции: массив из 4х эелементов
    # 1: число связных компонентов
    num_labels = output[0]
    # 2: изображение размером с исходное с 
    labels = output[1]
    # 3:
    stats = output[2]
    # 4:
    centroid = output[3]

    areas = [] # [stats[:, cv2.CC_STAT_AREA]]
    
    for i in range(num_labels):
        t = stats[i, cv2.CC_STAT_TOP   ]
        l = stats[i, cv2.CC_STAT_LEFT  ]
        w = stats[i, cv2.CC_STAT_WIDTH ]
        h = stats[i, cv2.CC_STAT_HEIGHT]
        a = stats[i, cv2.CC_STAT_AREA  ]

        if (a > 35 and a < 300):
            cv2.rectangle(frame, (l,t), (l+w,t+h), (123,34,234), 3)
            areas.append(a)
    
    # plt.hist(areas, bins=40)
    # plt.show()
    
    cv2.imshow("frame", frame)
    cv2.imshow("mask", mask)
    # cv2.imshow("")
    # cv2.imshow("labels", labels)

    key = cv2.waitKey(0) & 0xFF

    if (key == ord('q')):
        break

cam.release()
cv2.destroyAllWindows()
cv2.waitKey(10)
```
